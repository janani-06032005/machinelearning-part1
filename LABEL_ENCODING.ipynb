LABEL ENCODING

Converting label into numeric form

importing the dependicies

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

loading the data from csv files to pandas dataframe


```python
cancer_data=pd.read_csv('/content/breast_cancer_data.csv')
```

```python
cancer_data.head()
```

finding the count of different lables

```python
cancer_data['diagnosis'].value_counts()
```

LABEL ENCODING

```python
label_encode=LabelEncoder()
```

```python
labels=label_encode.fit_transform(cancer_data.diagnosis)
```

```python
cancer_data['target']=labels
```

```python
cancer_data.head()
```

```python
cancer_data['target'].value_counts()
```
